## 🚀 Step 4: Initialize TriageBot

## 💬 Step 5: Chat with TriageBot

Run this cell multiple times to have a conversation. Each time you run it, type a new message.

### 📋 Demo Messages (use these for your recording):

1. **FAQ Test:** `Why was I charged a $2 fee?`
2. **Account Lookup:** `What's the balance on account 4821?`
3. **Session Memory:** `And what was my last transaction?`
4. **Escalation:** `I want a $120 refund on this account.`

```python
# Get user input
user_message = input("\n🗣️  You: ")

if user_message.lower() in [':quit', ':exit', 'quit', 'exit']:
    print("👋 Goodbye!")
elif user_message.lower() == ':reset':
    bot.reset()
    print("🔄 Session memory cleared!")
else:
    # Get bot response
    print("\n🤖 TriageBot is thinking...\n")
    state = bot.respond(user_message)
    
    # Display the reply
    print("="*60)
    print(f"🤖 TriageBot: {state['final_reply']}")
    print("="*60)
    
    # Show trace details
    print("\n📊 How TriageBot handled this:")
    print(f"   • Branch taken: {state.get('route', 'N/A')}")
    
    if state.get('tool_calls'):
        print(f"   • Tools called: {len(state['tool_calls'])}")
        for i, tc in enumerate(state['tool_calls'], 1):
            print(f"     {i}. {tc.get('name', 'unknown')}")
    
    print(f"   • Redo attempts: {state.get('redo_count', 0)}")
    print(f"   • Review passed: {'✅ Yes' if state.get('review_passed') else '❌ No'}")
    
    if state.get('escalated'):
        print("   • ⚠️  ESCALATED TO HUMAN")
    
    if bot.last_account_id:
        print(f"   • 🧠 Remembering account: {bot.last_account_id}")
    
    print("\n" + "-"*60 + "\n")
```

## 🔄 Reset Session Memory

Run this to clear the bot's memory (useful between demo runs):

```python
bot.reset()
print("🔄 Session memory cleared! Ready for a fresh conversation.")
```
